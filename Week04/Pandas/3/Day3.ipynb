# Instructor Turn Activity 1: Merging 

```python
# Dependencies
import pandas as pd
```

```python
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["John", "Kelly", "Sam", "April", "Bobbo"],
    "email": ["jman@gmail", "kelly@aol.com", "sports@school.edu", "April@yahoo.com", "HeyImBobbo@msn.com"]
}
info_pd = pd.DataFrame(raw_data_info, columns=["customer_id", "name", "email"])
info_pd
```

```python
# Create DataFrames
raw_data_items = {
    "customer_id": [403, 112, 543, 999, 654],
    "item": ["soda", "chips", "TV", "Laptop", "Cooler"],
    "cost": [3.00, 4.50, 600, 900, 150]
}
items_pd = pd.DataFrame(raw_data_items, columns=[
                        "customer_id", "item", "cost"])
items_pd
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(info_pd, items_pd, on="customer_id")
merge_table
```

```python
# Merge two dataframes using an outer join
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="outer")
merge_table
```

```python
# Merge two dataframes using a left join
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="left")
merge_table
```

```python
# Merge two dataframes using a right join
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="right")
merge_table
```

## Students Turn Activity 2: Cryptocurrency Merging

### Instructions
  * Read in both of the CSV files and print out their DataFrames.

  * Perform an inner merge that combines both DataFrames on the "Date" column.

  * Rename the columns within the newly merged DataFrame so that the headers are more descriptive.

  * Create a summary table that includes the following information: `Best Bitcoin Open`, `Best Dash Open`, `Best Bitcoin Close`, `Best Dash Close`, `Total Bitcoin Volume`, `Total Dash Volume`.

  * `Total Bitcoin Volume` and `Total Dash Volume` should be calculated to have units of "millions" and be rounded to two decimal places.

```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged_crypto = pd.merge(bitcoin_df, dash_df, on="Date")
merged_crypto.head()
```

```python
# Rename columns so that they are differentiated
merged_crypto = merged_crypto.rename(columns={"Open_x":"Bitcoin_Open","High_x":"Bitcoin_High","Low_x":"Bitcoin_Low","Close_x":"Bitcoin_Close","Volume_x":"Bitcoin_Volume","Market Cap_x":"Bitcoin_Market_Cap",\
                                      "Open_y":"Dash_Open","High_y":"Dash_High","Low_y":"Dash_Low","Close_y":"Dash_Close","Volume_y":"Dash_Volume","Market Cap_y":"Dash_Market_Cap"})
merged_crypto.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
merged_crypto2 = pd.merge(bitcoin_df, dash_df, on="Date",suffixes=('_Bitcoin', '_Dash'))
merged_crypto2.head() 
```

```python
# Collecting best open for Bitcoin and Dash
bitcoin_best_open = merged_crypto["Bitcoin_Open"].max()
dash_best_open = merged_crypto["Dash_Open"].max()
# Collecting best close for Bitcoin and Dash
bitcoin_best_close = merged_crypto["Bitcoin_Close"].max()
dash_best_close = merged_crypto["Dash_Close"].max()

# Collecting the total volume for Bitcoin and Dash
bitcoin_volume = round(merged_crypto["Bitcoin_Volume"].sum()/1000000,2)
dash_volume = round(merged_crypto["Dash_Volume"].sum()/1000000,2)

#bitcoin_best_open
#dash_best_open
#bitcoin_best_close
#dash_best_close
#bitcoin_volume
dash_volume
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"Best Bitcoin Open":[bitcoin_best_open],
                          "Best Bitcoin Cloes":[bitcoin_best_close],
                          "Total Bitcoin Volume":f"{bitcoin_volume} millions",
                          "Best Dash Open":[dash_best_open],
                          "Best Dash Cloes":[dash_best_close],
                          "Total Dash Volume":f"{dash_volume} millions"})

summary_df.head()
```

# Instructor Turn Activity 3 Binning

```python
# Import Dependencies
import pandas as pd
```

```python
raw_data = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 56, 72, 88, 98, 67]}
df = pd.DataFrame(raw_data)
df
```

```python
# Create the bins in which Data will be held
# Bins are 0, 60, 70, 80, 90, 100
bins = [0, 60, 70, 80, 90, 100]

# Create the names for the four bins
group_names = ["F", "D", "C", "B", "A"]
```

```python
df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names)
df
```

```python
# Creating a group based off of the bins
df = df.groupby("Test Score Summary")
df.max()
```

## Students Turn Activity 4: Binning TED

### Instructions

  * Read in the CSV file provided and print it to the screen.

  * Find the minimum "views" and maximum "views".

  * Using the minimum and maximum "views" as a reference, create 10 bins in which to slice the data.

  * Create a new column called "View Group" and fill it with the values collected through your slicing.

  * Group the DataFrame based upon the values within "View Group".

  * Find out how many rows fall into each group before finding the averages for "comments", "duration", and "languages".

```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
highest_views = ted_df["views"].max()
lowest_views = ted_df["views"].min()

summary = [highest_views,lowest_views]

summary
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 10000000,20000000,30000000,40000000,50000000]
# Create labels for these bins
bin_labels = ["0 to 10M","10M to 20M","20M to 30M","30M to 40M", "40M+"]
```

```python
# Slice the data and place it into bins
ted_df["View Summary"] = pd.cut(ted_df["views"], bins, labels=bin_labels)
ted_df.head()
```

```python
# Place the data series into a new column inside of the DataFrame
ted_df["View Summary"] = pd.cut(ted_df["views"], bins, labels=bin_labels)
ted_df.head()
```

```python
# Create a GroupBy object based upon "View Group"
summary_data = ted_df.groupby("View Summary")

# Find how many rows fall into each bin
print(summary_data["comments"].count())

# Get the average of each column within the GroupBy object
summary_data[["comments","duration","languages"]].mean()
```

```python
# Import dependencies
import pandas as pd
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "Resources/EclipseBugs.csv"
eclipse_df = pd.read_csv(csv_path)
eclipse_df.head()
```

```python
# Finding the average number of comments per bug
eclipse_df.columns
# Finding the average number of comments per bug
average_comments = eclipse_df["Number of Comments"].mean()
average_comments
```

```python
# Grouping the DataFrame by "Assignee"
assignee_group = eclipse_df.groupby("Assignee")

# Count how many of each component Assignees worked on and create DataFrame
assignee_work = pd.DataFrame(assignee_group["Component"].value_counts())
assignee_work.head()
```

```python
# Rename the "Component" column to "Component Bug Count"
assignee_work = assignee_work.rename(columns={"Component":"Component Bug Count"})
assignee_work.head()
```

```python
# Find the percentage of bugs overall fixed by each Assignee
total_bugs = eclipse_df["Assignee"].count()
bugs_per_user = eclipse_df["Assignee"].value_counts()

#print(bugs_per_user)
user_bug_percent = pd.DataFrame((bugs_per_user/total_bugs)*100)
user_bug_percent.head()
```

```python
#rename the "assignee" column
user_bug_percent = user_bug_percent.rename(columns={"Assignee":"Percent of Total Bugs Assigned"})

#reset eht index
user_bug_percent = user_bug_percent.reset_index()
user_bug_percent.head()
```

```python
#Reset the index of "assignee_group"
assignee_work = assignee_work.reset_index()
assignee_work.head()

# Merge the "Percent of Total Bugs Assigned" into the DataFrame
assignee_work = assignee_work.merge(user_bug_percent, on="Assignee")
#assignee_work.head()
# Remove the extra columns
assignee_work = assignee_work[["Assignee", "Percent of Total Bugs", "Component", "Component Bug Count"]]
assignee_work.head()
```
