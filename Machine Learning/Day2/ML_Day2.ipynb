```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

df = pd.read_csv('LoansTrainingSet.csv')
```

```python
df = df[df["Current Loan Amount"] < 100000]
```

```python
df["Credit Score"] = df["Credit Score"].map(lambda x: x/10 if x > 800 else x)
```

```python
df.describe()
```

```python
df["Monthly Debt"] = df["Monthly Debt"].replace("[\$, ]","",regex = True).astype(float)
```

```python
df["Loan Status"] = df["Loan Status"].map(lambda x: 1 if x == "Fully Paid" else 0)
```

```python
df["Loan ID"].value_counts()
```

```python
df[df["Loan ID"] == "91343210-b13d-405c-b1a9-20429278208d"]
```

```python

#need to be improoved to keep the row with best data
df["Loan ID"].drop_duplicates(inplace=True)
```

```python
df["Home Ownership"].unique()
```

```python
df["Home Ownership"] = df["Home Ownership"].map(lambda x:"Home Mortgage" if x == "HaveMortgage" else x)
```

```python
df["Purpose"].unique()
```

```python
df["Purpose"] = df["Purpose"].map(lambda x:"Other" if x == "other" else x)
```

```python
df.drop(["Months since last delinquent"], axis=1, inplace=True)
```

```python
model_df = df[df["Credit Score"].notnull()]
```

```python
model_df.info()
```

```python
X = model_df.drop(["Loan ID", "Customer ID", "Maximum Open Credit", "Bankruptcies", "Tax Liens", "Loan Status", "Monthly Debt"], axis=1)
y = model_df["Loan Status"]
```

```python
X.info()
```

```python
X = pd.get_dummies(columns=["Term", "Years in current job", "Home Ownership", "Purpose"],data=X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# initialize model
log = LogisticRegression()
log.fit(X_train,y_train)

y_pred_log = log.predict(X_test)

acc_log = accuracy_score(y_test,y_pred_log)

print(acc_log)
```

```python
# Gradient Boosting (boosting)
params = {'n_estimators': 600, "max_depth" : 1, "learning_rate":0.1, "min_samples_leaf":1, "random_state": 42}
gbc = GradientBoostingClassifier(**params)
gbcmodel = gbc.fit(X_train, y_train)

y_predgbc = gbcmodel.predict(X_test)

acc_gbc = accuracy_score(y_predgbc,y_test)

print(acc_gbc)
```

```python
# Random Forest (bagging)
rf = RandomForestClassifier(n_estimators= 25, max_depth= None, max_features = 0.4, random_state = 42)
rfmodel = rf.fit(X_train, y_train)

y_predrf = rfmodel.predict(X_test)

acc_rf = accuracy_score(y_predrf, y_test)

print(acc_rf)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
# scaling methods: normalization (0 - 1), log
# (x - x_mean)/(x_std)
# zero mean, univariance (variance=1, std=1)
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
y_train_categorical
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
# hidden layer (1 layer)
model.add(Dense(units=6, activation='relu', input_dim=32)) 
# units is the number of nodes in the hidden layer
# input_dim is the number of features present in the input data

# output layer
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
# adam is a cathegorical alghoritm
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=32))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=50,
    shuffle=True,
    verbose=1
)
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
